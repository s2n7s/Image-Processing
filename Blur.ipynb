### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Read 'chess.jpg'

```python
chess = cv2.imread('chess.jpg')
```

### View the image with matplotlib

```python
plt.imshow(chess)
```

### Increase brightness by gamma correction

```python
def load_chess():
    chess = cv2.imread('chess.jpg').astype(np.float32)/255
    return chess
```

```python
def display_chess(chess):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.imshow(chess)
```

```python
chess = load_chess()
gamma = 1/100
effected_image = np.power(chess,gamma)
display_chess(effected_image)
```

### Apply Gaussian Blur on the original image

```python
chess = load_chess()
blurred_chess = cv2.GaussianBlur(chess,(5,5),10)
plt.imshow(blurred_chess)
```

### Apply median Blurring

```python
median = cv2.medianBlur(chess,5)
plt.imshow(median)
```

### Apply bilateral Filtering

```python
chess = load_chess()
display_chess(chess)
```
