### Import necessary Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Read 'image1.jpg' with opencv

```python
img = cv2.imread('image1.jpg')
```

```python
print(img)
```

### Plot the image with matplotlib

```python
plt.imshow(img)
```

### Convert the image to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

### Check the shape of the image

```python
img_rgb.shape
```

### Resize the image to (400,300)

```python
img_resize = cv2.resize(img_rgb, (400,300))
```

```python
plt.imshow(img_resize)
```

### Resize the image by ratio. Use 0.5 height and width

```python
img_resize = cv2.resize(img_rgb, (0,0), img, 0.5, 0.5)
```

```python
plt.imshow(img_resize)
```

### Flip the image along central x axis

```python
img_xaxis = cv2.flip(img_resize, 0)
plt.imshow(img_xaxis)
```

### Flip the image along central y axis 

```python
img_xaxis = cv2.flip(img_resize, 1)
plt.imshow(img_xaxis)
```
